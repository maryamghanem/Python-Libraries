**Import pandas and read in the banklist.csv file into a dataframe called banks.**

```python
import pandas as pd

# Load the data into a DataFrame
banks = pd.read_csv('banklist.csv')
```

```python
# 1. Show the head of the DataFrame

print(banks.head())
```

```python
# 2. What are the column names?

print(banks.columns)
```

```python
# 3. How many States (ST) are represented in this data set?

print(banks['ST'].nunique())
```

```python
# 4. Get a list or array of all the states in the data set.

print(banks['ST'].unique())
```

```python
# 5. What are the top 5 states with the most failed banks?

print(banks['ST'].value_counts().head(5))
```

```python
# 6. What are the top 5 acquiring institutions?

print(banks['Acquiring Institution'].value_counts().head(5))
```

```python
# 7. How many banks has the State Bank of Texas acquired?

state_bank_of_texas_acquisitions = banks[banks['Acquiring Institution'] == 'State Bank of Texas'].shape[0]
print(state_bank_of_texas_acquisitions)
```

```python
# 8. How many of them were actually in Texas?

texas_acquisitions = banks[(banks['Acquiring Institution'] == 'State Bank of Texas') & (banks['ST'] == 'TX')].shape[0]
print(texas_acquisitions)
```

```python
# 9. What is the most common city in California for a bank to fail in?

most_common_city_ca = banks[banks['ST'] == 'CA']['City'].value_counts().idxmax()
print(most_common_city_ca)
```

```python
# 10. How many failed banks don't have the word "Bank" in their name?

failed_banks_without_bank = banks[~banks['Bank Name'].str.contains('Bank', case=False)].shape[0]
print(failed_banks_without_bank)
```

```python
# 11. How many bank names start with the letter 's'?

banks_starting_with_s = banks[banks['Bank Name'].str.startswith('S')].shape[0]
print(banks_starting_with_s)
```

```python
# 12. How many CERT values are above 20000?

cert_values_above_20000 = banks[banks['CERT'] > 20000].shape[0]
print(cert_values_above_20000)
```

**Import pandas and read in the google-play-apps.csv file into a dataframe called apps_df.**

```python
import pandas as pd

# Load the data into a DataFrame
file_path = '/content/google-play-store.csv'
apps_df = pd.read_csv(file_path)

# Inspect the DataFrame
print(apps_df.head())
print(apps_df.columns)
```

```python
apps_df.info()
```

```python
# Explore the distribution of 'min_installs' to understand download trends

# Print some descriptive statistics
print(apps_df['min_installs'].describe())  # Shows stats like mean, median, min, max etc.

# Visualize the distribution using histograms
import matplotlib.pyplot as plt

plt.hist(apps_df['min_installs'])
plt.xlabel('Number of Installs')
plt.ylabel('Number of Apps')
plt.title('Distribution of Installs in Google Play Store Apps')
plt.show()

# You can interpret these results to answer questions 1 and 2:
# - Look for genres with high means or medians for categories with a lot of downloads.
# - Genres with low values or a concentration towards the lower end might have fewer downloads.
```

```python
# 1. What categories of applications get a lot of downloads per day?

# Clean the min_installs column and convert to integer
apps_df['min_installs'] = apps_df['min_installs'].replace(',', '').replace('+', '').astype(int)

# Convert the released column to datetime
apps_df['released'] = pd.to_datetime(apps_df['released'])

# Calculate the number of days since released
apps_df['Days Since Update'] = (pd.Timestamp('now') - apps_df['released']).dt.days

# Calculate downloads per day
apps_df['Downloads Per Day'] = apps_df['min_installs'] / apps_df['Days Since Update']

# Group by Category and get the average downloads per day
genre_grouped = apps_df.groupby('genre')
category_downloads = genre_grouped['Downloads Per Day'].mean().sort_values(ascending=False)

print(category_downloads.head(10))
```

```python
# 2. What categories of applications don't get many downloads per day?

print(category_downloads.tail(10))
```

```python
''' 3. In what app categories are there market leaders
(one app that clearly is getting downloaded more than the others)?'''

# Group the data by genre and find the app with the maximum number of min_installs
genre_leader = genre_grouped['min_installs'].idxmax()
print(genre_leader)
```

```python
# 4. How many downloads per day might you expect if you took the time to build an app?

average_downloads_per_day = apps_df['Downloads Per Day'].mean()
print(average_downloads_per_day)
```

```python
# 5. What can the data tell you about monetization approaches?

# Assuming monetization can be inferred from 'genre' and 'price' columns
monetization = genre_grouped[['price', 'min_installs']].mean().sort_values(by='price', ascending=False)
print(monetization.head(10))
```

```python
# 6. Calculate the average rating for each genre.

genre_avg_rating = genre_grouped['ratings'].mean().sort_values(ascending=False)
print(genre_avg_rating )
```

```python
# 7. Find the top 5 apps with the highest number of reviews.

# Replace non-numeric values in 'reviews' column with NaN
apps_df['reviews'] = pd.to_numeric(apps_df['reviews'], errors='coerce')

# Drop rows with NaN values in 'reviews' column
apps_df.dropna(subset=['reviews'], inplace=True)

apps_df['reviews'] = apps_df['reviews'].astype(int)
top_5_apps_by_reviews = apps_df.sort_values(by='reviews', ascending=False).head(5)
print(top_5_apps_by_reviews[['title', 'reviews']])
```

```python
# 8.Count the number of apps that offer in-app purchases (IAP).

# Count the number of apps that have 'True' in the 'offers_iap' column
num_iap_apps = apps_df[apps_df['offers_iap']].shape[0]
print(f"Number of apps that offer in-app purchases: {num_iap_apps}")
```

```python
# 9. Determine the total number of installs for each genre.

# Calculate total installs for each genre, reset index to get a DataFrame
genre_total_installs = genre_grouped['min_installs'].sum().reset_index()
print(genre_total_installs)
```

```python
# 10. Find the earliest and latest release date in the dataset.

earliest_release_date = apps_df['released'].min()
latest_release_date = apps_df['released'].max()
print(earliest_release_date, latest_release_date)
```

```python
# 11. Calculate the average number of ratings per day for apps with more than 10,000 reviews.

filtered_df = apps_df[apps_df['reviews'] > 10000]
avg_ratings_per_day = filtered_df['ratings_per_day'].mean()
avg_ratings_per_day
print(avg_ratings_per_day)
```

```python
# 12. Identify the top 3 genres with the highest average app score.

top_3_genres = genre_avg_rating.nlargest(3)
print(top_3_genres)
```

```python
# 13. Determine the percentage of apps that are ad-supported and offer in-app purchases.

# Calculate the percentage
ad_supported_percentage = apps_df['ad_supported'].mean() * 100
iap_percentage = apps_df['offers_iap'].mean() * 100
print(f"Percentage of apps that are ad-supported: {ad_supported_percentage}%")
print(f"Percentage of apps that offer in-app purchases: {iap_percentage}%")
```

```python
# 14. Find the most common release year for apps in the dataset.

# Convert 'released' column to datetime if it's not already
apps_df['released'] = pd.to_datetime(apps_df['released'])

apps_df['Release Year'] = apps_df['released'].dt.year
most_common_release_year = apps_df['Release Year'].mode()[0]
print(most_common_release_year)
```

```python
# 15. Calculate the correlation between the number of reviews and the average rating for each genre.

correlation_reviews_rating = genre_grouped.apply(lambda x: x['reviews'].corr(x['ratings']))
print(correlation_reviews_rating)
```

```python
# 16. Calculate the weighted average rating for each genre, where the weight is the number of reviews.

weighted_average_rating = genre_grouped.apply(lambda x: (x['ratings'] * x['reviews']).sum() / x['reviews'].sum()).sort_values(ascending=False)
print(weighted_average_rating)
```

```python
# 17. Identify the app with the highest ratio of ratings per review for each genre.

apps_df['ratings_per_review'] = apps_df['ratings'] / apps_df['reviews']
highest_ratings_per_review_per_genre = apps_df.loc[apps_df.groupby('genre')['ratings_per_review'].idxmax()]
highest_ratings_per_review_per_genre[['genre', 'title', 'ratings_per_review']].head()
```

**Import pandas and read in the google-play-apps.csv file into a dataframe called apps_df.**

```python
# Step 1: Import Libraries and Read Data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objs as go

# Read the CSV file into a DataFrame
apps_df = pd.read_csv('google-play-store.csv')
```

```python
# 1. Create a simple bar plot using Matplotlib to show the count of apps in each genre.

plt.figure(figsize=(10,6))
apps_df['genre'].value_counts().plot(kind='bar')
plt.title('Count of Apps in Each Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Using Seaborn, create a histogram to show the distribution of app ratings.

plt.figure(figsize=(12, 8))
sns.histplot(apps_df['score'], bins=30, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Generate a scatter plot using Matplotlib to visualize the relationship between app size and number of reviews.

# Unfortunately 'app size' coulmn not found :(
```

```python
# 4. Create a Seaborn count plot to show the number of free and paid apps.

apps_df['free'] = apps_df['price'] == 0.0
plt.figure(figsize=(8, 6))
sns.countplot(data=apps_df, x='free')
plt.title('Number of Free Vs Paid Apps')
plt.xlabel('Type')
plt.ylabel('Number Of Apps')
plt.xticks([0, 1], ['Paid', 'Free'])
plt.show()
```

```python
# 5. Using Plotly, create an interactive line plot showing the trend of the average app rating over time.

apps_df['released'] = pd.to_datetime(apps_df['released'])
apps_df['Year'] = apps_df['released'].dt.year
avg_rating_by_year = apps_df.groupby('Year')['ratings'].mean().reset_index()

fig = px.line(avg_rating_by_year, x='Year', y='ratings', title='Trend of Average App Rating Over Time')
fig.show()
```

```python
# 6. Using Plotly, create an interactive bar plot to show the average number of installs for free vs. paid apps.

apps_df['min_installs'] = apps_df['min_installs'].replace('[+,]', '').astype(int)
avg_installs_by_type = apps_df.groupby('offers_iap')['min_installs'].mean().reset_index()

fig = px.bar(avg_installs_by_type, x='offers_iap', y='min_installs', title='Average Number of Installs for Free vs. Paid Apps')
fig.show()
```

```python
# 7. Create a Matplotlib pie chart to show the proportion of apps that are ad-supported versus those that are not.

ad_supported_count = apps_df['ad_supported'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(ad_supported_count, labels=['Ad-Supported', 'Non-Ad-Supported'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Ad-Supported vs. Non-Ad-Supported Apps')
plt.show()
```

```python
# 8. Using Seaborn, create a pairplot to analyze the relationships between reviews, ratings, and price for the top 5 genres with the most apps.

top_genres = apps_df['genre'].value_counts().head(5).index.tolist()
top_genres_df = apps_df[apps_df['genre'].isin(top_genres)]

sns.pairplot(top_genres_df, hue='genre', vars=['reviews', 'score', 'price'], palette='Set1')
plt.suptitle('Pairplot for Reviews, Ratings, and Price by Genre', y=1.02)
plt.show()
```

```python
# Bonus:
# 1. Create an interactive Plotly box plot to visualize the distribution of app prices across different genres. Add a dropdown menu to filter the genres.

fig = px.box(apps_df, x='genre', y='price', points='all')
fig.update_layout(
    title='Distribution of App Prices Across Different Genres',
    xaxis_title='Genre',
    yaxis_title='Price',
    updatemenus=[
        {
            'buttons': [
                {
                    'method': 'restyle',
                    'label': genre,
                    'args': [{'x': [apps_df[apps_df['genre'] == genre]['genre']],
                              'y': [apps_df[apps_df['genre'] == genre]['price']]}]
                } for genre in apps_df['genre'].unique()
            ],
            'direction': 'down',
        }
    ]
)
fig.show()
```

```python
#Bonus:
# 2. Using Seaborn, create a violin plot to compare the distribution of ratings for free versus paid apps.

plt.figure(figsize=(10,6))
sns.violinplot(x='offers_iap', y='ratings', data=apps_df)
plt.title('Distribution of Ratings for Free vs. Paid Apps')
plt.xlabel('Type')
plt.ylabel('Rating')
plt.xticks([0, 1], ['Paid', 'Free'])
plt.show()
```
